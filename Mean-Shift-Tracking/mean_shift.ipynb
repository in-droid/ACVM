```python
import numpy as np
import matplotlib.pyplot as plt

from ex2_utils import generate_responses_1
from ex1_utils import gausssmooth
import math

from mean_shift import mean_shift
```

```python
def generate_custom():
    x = np.zeros((100, 100))
    x[10, 10] = 10
    x[20, 80] = 150
    x[50, 70] = 90
    x[90, 90] = 40
    return gausssmooth(x, 10)
```

```python
I = generate_custom()
h = 19
pos = (50, 50)
res_uniform = mean_shift(I, pos, h, return_path=True)
res_normal = mean_shift(I, pos, 17, kernel_type='normal', return_path=True)


plt.imshow(I, cmap='gray')
print("Starting position:", pos)
circle = plt.Circle(pos, h//2, color='r', fill=False, alpha=1)
plt.gca().add_patch(circle)
for pos in res_uniform[:-1]:
    circle = plt.Circle(pos, h//2, color='r', fill=False, alpha=.2)
    plt.gca().add_patch(circle)

for pos in res_normal[:-1]:
    circle = plt.Circle(pos, h//2, color='g', fill=False, alpha=.2)
    plt.gca().add_patch(circle)

print("Number of iterations (uniform):", len(res_uniform)-1)
print("Final position (uniform):", res_uniform[-1])
print("Function value (uniform):", I[res_uniform[-1][1], res_uniform[-1][0]])

print("Number of iterations (normal):", len(res_normal)-1)
print("Final position (normal):", res_normal[-1])
print("Function value (normal):", I[res_normal[-1][1], res_normal[-1][0]])


circle = plt.Circle(res_uniform[-1], h//2, color='r', fill=False, alpha=1)
plt.gca().add_patch(circle)
circle = plt.Circle(res_normal[-1], h//2, color='g', fill=False, alpha=1)
plt.gca().add_patch(circle)

plt.imshow(I, 'gray')
# plt.show()
plt.savefig('./data/plots/generate_responses_custom.pdf')
```
